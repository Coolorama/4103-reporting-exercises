# A. Nonlinear Transformation

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load the CSV file
file_path = '1_china_gdp.csv'
data = pd.read_csv(file_path)

# Display the first few rows of the dataset
data.head()

# Perform a nonlinear transformation (quadratic function)
original_column = 'Value'
transformed_column = f'{original_column}_squared'

data[transformed_column] = data[original_column] ** 2

# Display the updated DataFrame
data.head()

# Scatter plot for visualization
plt.figure(figsize=(10, 6))

# Use "year" and "value_squared" columns for the scatter plot
plt.scatter(data['Year'], data[transformed_column], label='Transformed Data', color='blue')
plt.xlabel('Year')
plt.ylabel('Transformed Data')
plt.title('Nonlinear Transformation: Quadratic Function')
plt.legend()
plt.grid(True)
plt.show()
```

# B. Categorical Encoding

```python
import pandas as pd

# Load the CSV file
file_path = '2_melb_data.csv'
data = pd.read_csv(file_path)

# Display the first few rows of the dataset
data.head()

print("Unique values in 'Type' column:", data['Type'].unique())

# Perform one-hot encoding on the 'Type' column
data_encoded = pd.get_dummies(data, columns=['Type'], prefix='Type')

# Display the updated DataFrame with one-hot encoding
data_encoded.head()
```

# C. Discretization

```python
import pandas as pd

# Load the CSV file
file_path = '3_train.csv'
data = pd.read_csv(file_path)

# Display the first few rows of the dataset
data.head()

continuous_variable = 'Age'

# Perform equal-width discretization
num_bins = 5  # You can adjust the number of bins based on your preference
data['Age_Discretized'] = pd.cut(data[continuous_variable], bins=num_bins, labels=False)

# Display the updated DataFrame with the discretized column
data.head()
```
